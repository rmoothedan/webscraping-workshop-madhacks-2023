```python
import requests
```

```python
url = "https://davidteather.github.io/webscraping-workshop-website-madhacks-2023/"
r = requests.get(url)
r
```

```python
r.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
for product in soup.find_all('div', attrs={ 'class': 'product-details'}):
    print(product)
    print('\n')
```

```python
for product in soup.find_all('div', attrs={ 'class': 'product-details'}):
    # print(product.p)
    # for child in product.final_all('p'):
    #     print(child)

    children = product.findChildren('p')
    price = children[1].get_text()
    print(price)
```

```python
for product in soup.find_all('div', attrs={ 'class': 'product-details'}):
    print(product.h4.get_text(), product.findChildren('p')[1].get_text())
```
